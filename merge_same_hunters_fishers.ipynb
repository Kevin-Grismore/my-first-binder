```python
import pandas as pd
import prep

df = prep.prep_all_states()
```

```python
df.shape
```

```python
df.head()
```

```python
hunt_df = df[df['Hunt'] == 'Y'].drop(columns=['Fish'])
hunt_df.head()
```

```python
fish_df = df[df['Fish'] == 'Y'].drop(columns=['Hunt'])
fish_df.head()
```

```python
columns = list(df.columns)
columns.remove('Hunt')
columns.remove('Fish')

df = hunt_df.merge(fish_df, how='outer', on=columns)
df.shape
```

```python
df.head()
```

```python
df['RecordID'] = df.index + 1
df['Addr2'] = ''
df = df.reindex(columns=['RecordID', 'FirstName', 'MiddleName', 'LastName', 'Suffix', 'Street', 'Addr2', 'City', 'State', 'Zip', 'Hunt', 'Fish'])
df = df.fillna('')
```

```python
df.to_csv('hunt_fish_out.csv', index=False)
```
